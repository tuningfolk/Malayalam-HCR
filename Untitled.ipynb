```python
import numpy as np
import cv2
import random

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Sample-handwritten-text-input-for-OCR.png')
print('image shape: {}'.format(img.shape))
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
plt.show()
```

```python
ret, img_thresh = cv2.threshold(img_gray, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(img_thresh)
plt.show()
```

```python
cv2.imshow('Otsu Threshold', img_thresh)
```
